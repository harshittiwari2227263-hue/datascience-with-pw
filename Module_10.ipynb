# 1. What is a random variable in probability theory?

A random variable is a numerical value that represents the outcome of a random experiment.

# 2. What are the types of random variables?

There are two main types: discrete random variables and continuous random variables.

# 3. What is the difference between discrete and continuous distributions?

Discrete distributions deal with countable values, while continuous distributions deal with values in a continuous range.

# 4. What are probability distribution functions (PDF)?

A PDF describes how probabilities are distributed over values of a continuous random variable.

# 5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

A CDF gives the probability that a random variable is less than or equal to a value, while a PDF shows the density of probability at each point.

# 6. What is a discrete uniform distribution?

It is a distribution where all possible discrete outcomes have equal probability.

# 7. What are the key properties of a Bernoulli distribution?

It models a single trial with two outcomes (success or failure) and is defined by probability p of success.

# 8. What is the binomial distribution, and how is it used in probability?

It models the number of successes in n independent Bernoulli trials with probability p.

# 9. What is the Poisson distribution and where is it applied?

It models the number of events occurring in a fixed interval of time or space when events happen independently at a constant rate.

# 10. What is a continuous uniform distribution?

It is a distribution where all values within a given continuous interval have equal probability density.

# 11. What are the characteristics of a normal distribution?

It is symmetric, bell-shaped, defined by mean (μ) and standard deviation (σ), and total area equals 1.

# 12. What is the standard normal distribution, and why is it important?

It is a normal distribution with mean 0 and standard deviation 1, used to calculate probabilities using Z-scores.

# 13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

CLT states that the sampling distribution of the sample mean approaches normal distribution as sample size increases, regardless of population shape.

# 14. How does the Central Limit Theorem relate to the normal distribution?

It explains why sample means tend to follow a normal distribution for large samples.

# 15. What is the application of Z statistics in hypothesis testing?

Z-statistics are used to test hypotheses about population means when variance is known or sample size is large.

# 16. How do you calculate a Z-score, and what does it represent?

Z = (X − μ) / σ; it represents how many standard deviations a value is from the mean.

# 17. What are point estimates and interval estimates in statistics?

A point estimate gives a single value estimate of a parameter, while an interval estimate gives a range of values.

# 18. What is the significance of confidence intervals in statistical analysis?

They provide a range within which the population parameter is likely to lie with a certain confidence level.

# 19. What is the relationship between a Z-score and a confidence interval?

Confidence intervals are calculated using Z-scores to determine margin of error.

# 20. How are Z-scores used to compare different distributions?

They standardize values so comparisons can be made across different scales.

# 21. What are the assumptions for applying the Central Limit Theorem?

Random sampling, independence of observations, and sufficiently large sample size.

# 22. What is the concept of expected value in a probability distribution?

Expected value is the weighted average of all possible outcomes of a random variable.
