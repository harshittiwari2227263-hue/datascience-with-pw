```python
# 1.What is statistics, and why is it important?
# ANS: Statistics is the science of collecting, analyzing, interpreting, presenting, and organizing data. It is important because it helps us make informed decisions based on data, identify trends and patterns, and draw conclusions about populations from samples. Statistics is widely used in various fields such as business, healthcare, social sciences, and many others to understand and solve real-world problems.

# 2.What are the two main types of statistics?
# ANS: The two main types of statistics are descriptive statistics and inferential statistics. Descriptive statistics summarize and describe the main features of a dataset, while inferential statistics use sample data to make inferences or predictions about a larger population.

# 3.What are descriptive statistics?
# ANS: Descriptive statistics are methods for summarizing and describing the main features of a dataset. They include measures of central tendency (mean, median, mode), measures of variability (range, variance, standard deviation), and measures of distribution (skewness, kurtosis). Descriptive statistics help us understand the basic characteristics of the data and provide a simple summary of the sample and the measures.

# 4.What is inferential statistics
# ANS: Inferential statistics is a branch of statistics that allows us to make inferences or predictions about a population based on a sample of data. It involves using probability theory to estimate population parameters, test hypotheses, and make predictions. Inferential statistics helps us draw conclusions about a larger group based on the analysis of a smaller subset of data.

# 5.What is sampling in statistics?
# ANS: Sampling in statistics is the process of selecting a subset of individuals, items, or data points from a larger population to represent that population. The goal of sampling is to gather information about the population without having to collect data from every member. Proper sampling techniques are crucial for ensuring that the sample is representative of the population, which allows for accurate inferences and conclusions to be drawn from the data.

# 6.What are the different types of sampling methods?
# ANS: There are several types of sampling methods, including:
# 1. Simple Random Sampling: Every member of the population has an equal chance of being selected.
# 2. Systematic Sampling: Every nth member of the population is selected after a random starting point.
# 3. Stratified Sampling: The population is divided into subgroups (strata) based on certain characteristics, and random samples are taken from each stratum.
# 4. Cluster Sampling: The population is divided into clusters, and a random sample of clusters is selected. All members of the selected clusters are included in the sample.
# 5. Convenience Sampling: Samples are taken from a group that is conveniently accessible to the researcher, which may lead to bias.

# 7.What is the difference between random and non-random sampling?
# ANS: Random sampling is a method where every member of the population has an equal chance of being selected for the sample. Non-random sampling is a method where members are selected based on convenience, judgment, or other non-random criteria, which may introduce bias.

# 8.Define and give examples of qualitative and quantitative data?
# ANS: Qualitative data is non-numeric and describes qualities or characteristics. Examples include colors, names, and categories. Quantitative data is numeric and can be measured or counted. Examples include height, weight, and age.

# 9.What are the different types of data in statistics?
# ANS: The different types of data in statistics include:
# 1. Nominal Data: Categorical data without a specific order (e.g., gender, colors).
# 2. Ordinal Data: Categorical data with a specific order (e.g., rankings, education levels).
# 3. Interval Data: Numeric data without a true zero point (e.g., temperature in Celsius).
# 4. Ratio Data: Numeric data with a true zero point (e.g., weight, height).

# 10.Explain nominal, ordinal, interval, and ratio levels of measurement?
# ANS:
# 1. Nominal Level: This level of measurement categorizes data without any order or hierarchy. Examples include gender, colors, and types of animals.
# 2. Ordinal Level: This level of measurement categorizes data with a specific order or ranking. Examples include education levels (high school, bachelor's, master's) and customer satisfaction ratings (satisfied, neutral, dissatisfied).
# 3. Interval Level: This level of measurement involves numeric data where the intervals between values are meaningful, but there is no true zero point. Examples include temperature in Celsius or Fahrenheit.
# 4. Ratio Level: This level of measurement involves numeric data with a true zero point, allowing for meaningful comparisons of ratios. Examples include weight, height, and income.

# 11.What is the measure of central tendency?
# ANS: The measure of central tendency is a statistical measure that describes the center or typical value of a dataset. The three main measures of central tendency are mean (average), median (the middle value when data is ordered), and mode (the most frequently occurring value). These measures help summarize and describe the main characteristics of a dataset.

# 12.Define mean, median, and mode?
# ANS:
# 1. Mean: The arithmetic average of a dataset, calculated by summing all values and dividing by the number of values.
# 2. Median: The middle value in a dataset when the values are arranged in ascending or descending order.
# 3. Mode: The value that appears most frequently in a dataset.

# 13.What is the significance of the measure of central tendency?
# ANS: The measure of central tendency is significant because it provides a summary statistic that represents the center of a dataset. It helps us understand the typical value or average of the data, which can be useful for making comparisons, identifying trends, and making informed decisions based on the data. Different measures of central tendency may be more appropriate depending on the distribution and nature of the data.

# 14.What is variance, and how is it calculated?
# ANS: Variance is a measure of how much the values in a dataset vary or deviate from the mean. It is calculated by taking the average of the squared differences between each value and the mean. The formula for variance (σ²) is:
# σ² = (Σ (x_i - μ)²) / N

# 15.What is skewness in a dataset?
# ANS: Skewness is a measure of the asymmetry of a probability distribution or dataset. It indicates whether the data is skewed to the left (negatively skewed) or to the right (positively skewed). A skewness value of zero indicates a perfectly symmetrical distribution, while positive skewness indicates a longer tail on the right side and negative skewness indicates a longer tail on the left side. Skewness helps us understand the shape of the data distribution and can impact statistical analyses and interpretations.

# 16.What is standard deviation, and why is it important?
# ANS: Standard deviation is a measure of the amount of variation or dispersion in a dataset. It is calculated as the square root of the variance. Standard deviation is important because it provides insight into how spread out the values in a dataset are around the mean. A low standard deviation indicates that the values are close to the mean, while a high standard deviation indicates that the values are more spread out. Standard deviation is widely used in statistics to assess the variability of data and to make comparisons between different datasets.

# 17.Define and explain the term range in statistics?
# ANS: The range in statistics is a measure of the spread or dispersion of a dataset. It is calculated as the difference between the maximum and minimum values in the dataset. The formula for range is:
# Range = Maximum value - Minimum value

# 18.What is the difference between variance and standard deviation
# ANS: Variance and standard deviation are both measures of variability in a dataset. Variance is the average of the squared differences from the mean, while standard deviation is the square root of the variance. The main difference is that standard deviation is expressed in the same units as the original data, making it more interpretable, while variance is expressed in squared units, which can be less intuitive to understand. Standard deviation is often preferred for reporting variability because it provides a clearer picture of how much the data deviates from the mean.

# 19.What does it mean if a dataset is positively or negatively skewed?
# ANS: If a dataset is positively skewed, it means that the tail of the distribution is longer on the right side, indicating that there are more values that are higher than the mean. In contrast, if a dataset is negatively skewed, it means that the tail of the distribution is longer on the left side, indicating that there are more values that are lower than the mean. Skewness can affect statistical analyses and interpretations, as it may indicate that the data does not follow a normal distribution and may require different analytical approaches.

# 20.Define and explain kurtosis?
# ANS: Kurtosis is a statistical measure that describes the shape of a probability distribution, specifically the "tailedness" or the presence of outliers in the data. A distribution with high kurtosis has heavy tails and more extreme outliers, while a distribution with low kurtosis has light tails and fewer outliers. The formula for kurtosis is:
# Kurtosis = (Σ (x_i - μ)⁴) / (N * σ⁴)
# A normal distribution has a kurtosis of 3, and excess kurtosis is calculated by subtracting 3 from the kurtosis value. Positive excess kurtosis indicates a distribution with heavier tails than a normal distribution, while negative excess kurtosis indicates a distribution with lighter tails. Kurtosis helps us understand the likelihood of extreme values in the dataset and can impact statistical analyses and interpretations.

# 21.What is the purpose of covariance?
# ANS: Covariance is a measure of how two random variables change together. It indicates the direction of the linear relationship between the variables. A positive covariance means that as one variable increases, the other variable tends to increase as well, while a negative covariance means that as one variable increases, the other variable tends to decrease. Covariance is used in statistics to assess the relationship between two variables and is a key component in calculating correlation, which standardizes the covariance to provide a more interpretable measure of the strength and direction of the relationship.

# 22.What does correlation measure in statistics?
# ANS: Correlation measures the strength and direction of the linear relationship between two variables. It is a standardized version of covariance that ranges from -1 to 1. A correlation of 1 indicates a perfect positive linear relationship, a correlation of -1 indicates a perfect negative linear relationship, and a correlation of 0 indicates no linear relationship between the variables. Correlation is widely used in statistics to assess the degree to which two variables are related and to make predictions based on that relationship.

# 23.What is the difference between covariance and correlation?
# ANS: The main difference between covariance and correlation is that covariance is a measure of how two variables change together, while correlation is a standardized measure of the strength and direction of the linear relationship between two variables. Covariance can take any value and is not standardized, making it difficult to interpret, while correlation ranges from -1 to 1, providing a clear indication of the strength and direction of the relationship. Correlation is often preferred for reporting relationships between variables because it is more interpretable and allows for easier comparisons across different datasets.

# 24.What are some real-world applications of statistics?
# ANS: Statistics has numerous real-world applications across various fields, including:
# 1. Healthcare: Statistics is used to analyze medical data, evaluate treatment effectiveness, and conduct clinical trials.
# 2. Business: Statistics is used for market research, quality control, and financial analysis.
# 3. Social Sciences: Statistics is used to analyze survey data, study social trends, and conduct experiments.
# 4. Sports: Statistics is used to analyze player performance, team strategies, and game outcomes.
# 5. Government: Statistics is used for census data analysis, policy evaluation, and economic forecasting.
# 6. Education: Statistics is used to analyze student performance, evaluate teaching methods, and conduct educational research.
```

```python
# 1. How do you calculate the mean, median, and mode of a dataset?
import statistics as stats

data = [10, 20, 20, 30, 40, 50, 50, 50]

mean_val = sum(data) / len(data)
median_val = stats.median(data)
mode_val = stats.mode(data)

print("Data:", data)
print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
```

```python
# 2. Write a Python program to compute the variance and standard deviation of a dataset?
import numpy as np

data = [10, 20, 20, 30, 40, 50, 50, 50]
arr = np.array(data)

variance_pop = np.var(arr)          # population variance
std_pop = np.std(arr)               # population std

variance_sample = np.var(arr, ddof=1)  # sample variance
std_sample = np.std(arr, ddof=1)       # sample std

print("Population Variance:", variance_pop)
print("Population Std Dev:", std_pop)
print("Sample Variance:", variance_sample)
print("Sample Std Dev:", std_sample)
```

```python
# 4. Implement sampling techniques like random sampling and stratified sampling?
import pandas as pd

df = pd.DataFrame({
    "Name": [f"S{i}" for i in range(1, 21)],
    "Group": ["A"]*10 + ["B"]*10,
    "Score": [45,50,60,65,70,55,58,62,67,72, 40,42,48,52,57,59,61,64,68,75]
})
# Random Sampling
random_sample = df.sample(n=5, random_state=42)

# Stratified Sampling (same proportion from each group)
stratified_sample = df.groupby("Group", group_keys=False).apply(lambda x: x.sample(2, random_state=42))
print("Random Sample:\n", random_sample)
print("\nStratified Sample:\n", stratified_sample)
```

```python
# 8. Generate a dataset and demonstrate positive and negative skewness?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

positive_skew = np.random.exponential(scale=2.0, size=1000)     # right-skew
negative_skew = -np.random.exponential(scale=2.0, size=1000)    # left-skew

print("Positive skew value:", skew(positive_skew))
print("Negative skew value:", skew(negative_skew))

plt.hist(positive_skew, bins=30)
plt.title("Positive (Right) Skew")
plt.show()

plt.hist(negative_skew, bins=30)
plt.title("Negative (Left) Skew")
plt.show()
```

```python
# 9. Write a Python script to calculate covariance between two datasets?
import numpy as np

x = np.array([10, 20, 30, 40, 50])
y = np.array([15, 25, 35, 45, 60])

cov_matrix = np.cov(x, y, ddof=1)  # sample covariance matrix
cov_xy = cov_matrix[0, 1]

print("Covariance Matrix:\n", cov_matrix)
print("Cov(x,y):", cov_xy)
```

```python
# 10. Write a Python script to calculate the correlation coefficient between two datasets?
import numpy as np
x = np.array([10, 20, 30, 40, 50])
y = np.array([15, 25, 35, 45, 60])
corr_matrix = np.corrcoef(x, y)
corr_xy = corr_matrix[0, 1]
print("Correlation Matrix:\n", corr_matrix)
print("Correlation(x,y):", corr_xy)
```

```python
# 12. Implement and compare simple random sampling and systematic sampling?
import pandas as pd
import numpy as np

df = pd.DataFrame({"Value": range(1, 51)})

# Simple Random Sampling
srs = df.sample(n=10, random_state=1)

# Systematic Sampling (every k-th element)
k = len(df) // 10
start = 0
systematic = df.iloc[start::k].head(10)

print("Simple Random Sample:\n", srs.T)
print("\nSystematic Sample:\n", systematic.T)
```

```python
#17. Calculate the interquartile range (IQR) of a dataset?
data = np.array([10, 12, 15, 18, 20, 22, 25, 30, 35, 40])
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

print("Q1:", q1)
print("Q3:", q3)
print("IQR:", iqr)
```

```python
# 18. Implement Z-score normalization and explain its significance?
import numpy as np

data = np.array([10, 20, 30, 40, 50])

z = (data - np.mean(data)) / np.std(data, ddof=0)

print("Original:", data)
print("Z-score normalized:", z)
```

```python
# 20. Write a Python program to visualize covariance using a heatmap?
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Create dataset
df = pd.DataFrame({
    "X": np.random.normal(0, 1, 200),
    "Y": np.random.normal(0, 1, 200),
    "Z": np.random.normal(0, 1, 200)
})

# Covariance matrix
cov = df.cov()

# Heatmap with values
sns.heatmap(cov, annot=True, fmt=".2f")
plt.title("Covariance Heatmap")
plt.show()

```

```python
# 21. Generate a dataset and implement both variance and standard deviation computations?

data = np.random.randint(1, 101, 20)

mean = np.mean(data)
variance_manual = np.mean((data - mean) ** 2)     # population variance
std_manual = np.sqrt(variance_manual)

print("Data:", data)
print("Manual Variance:", variance_manual)
print("Manual Std Dev:", std_manual)

print("NumPy Variance:", np.var(data))
print("NumPy Std Dev:", np.std(data))
```

```python
# 23. Implement the Pearson and Spearman correlation coefficients for a dataset?

import numpy as np
from scipy.stats import pearsonr, spearmanr

x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([2, 4, 5, 4, 5, 7])

pearson_corr, _ = pearsonr(x, y)
spearman_corr, _ = spearmanr(x, y)

print("Pearson Correlation:", pearson_corr)
print("Spearman Correlation:", spearman_corr)
```
